```python
def FACE_DETECTION_MTCNN(image):
  print("[INFO] Processing FACE_DETECTION_CNN")
  from mtcnn import MTCNN
  detector = MTCNN()
  faces = detector.detect_faces(image)

  for result in faces:
    x,y,w,h = result['box']
    x1,y1 = x+w,y+h
    cv.rectangle(image,(x,y),(x1,y1),(0,0,255),2)
  print('[INFO] Processing Done')
  return image
  
```

```python
image_done = FACE_DETECTION_MTCNN(img)
cv_imshow(image_done)
```
