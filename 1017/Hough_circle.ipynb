```python
import cv2 as cv
import os
import sys
import numpy as np
from matplotlib import pyplot as plt
from google.colab import drive
from google.colab.patches import cv_imshow
drive.mount('/content/drive')
```

```python
folder = r'/content/drive/MyDrive/images'
path_img = os.path.join(folder,'NTUST_logo.png')

img = cv.imread(path_img)
if img is None:
  sys.exit("could not read the image.")
cv_imshow(img)
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
def Edge_Detection_HoughCircle(original_image,image):
  copy_image=original_image.copy()
  # edges = cv.Canny(copy_image,50,200)
  circles=cv.HoughCircles(image,cv.HOUGH_GRADIENT,1,50,0,10,minRadius=0,maxRadius=0)
  circles=np.uint16(np.around(circles))

  for i in circles[0,:]:
    cv.circle(copy_image,(i[0],i[1]),i[2],(0,255,0),2)
    cv.circle(copy_image,(i[0],i[1]),2,(0,0,2),3)
  return copy_image
```

```python
img_circle = Edge_Detection_HoughCircle(img,img_gray)
cv_imshow(img)
cv_imshow(img_circle)
```
