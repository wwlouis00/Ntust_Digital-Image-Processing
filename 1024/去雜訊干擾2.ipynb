```python
import cv2 as cv
import os
import sys
import numpy as np
from matplotlib import pyplot as plt
from google.colab import drive
from google.colab.patches import cv_imshow
drive.mount('/content/drive')
```

```python
folder = r'/content/drive/MyDrive/images'
path_img = os.path.join(folder,'Number_noise.png')

img = cv.imread(path_img)
if img is None:
  sys.exit("could not read the image.")
cv_imshow(img)
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

# Morphological Transformations: Erosion(侵蝕)

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img_gray,kernel,iterations = 1)

res = np.hstack((img_gray,erosion))
cv_imshow(res)
```

# Morphological Transformations: Dilation (膨脹)


```python
kernel = np.ones((5,5),np.uint8)
dilation = cv.dilate(img_gray,kernel,iterations = 1)

res = np.hstack((img_gray,dilation))
cv_imshow(res)
```

# Morphological Transformations: Opening (斷開/開運算)


```python
kernel = np.ones((5,5),np.uint8)
opening= cv.morphologyEx(img_gray,cv.MORPH_OPEN,kernel)

res = np.hstack((img_gray,opening))
cv_imshow(res)
```

# Morphological Transformations: Closing (閉合/閉運算)


```python
kernel = np.ones((5,5),np.uint8)
closing= cv.morphologyEx(img_gray,cv.MORPH_CLOSE,kernel)

res = np.hstack((img_gray,closing))
cv_imshow(res)
```
